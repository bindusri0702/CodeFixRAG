```python
from langchain.text_splitter import TokenTextSplitter
text_splitter = TokenTextSplitter(
    chunk_size=450, # 500 tokens is the max
    chunk_overlap=20 # Overlap of N tokens between chunks (to reduce chance of cutting out relevant connected text like middle of sentence)
)
```

```python
from langchain.docstore.document import Document
```

```python
documents = []
chunks = text_splitter.split_text(code_texts)
total_chunks = len(chunks)
for chunk_num in range(1,total_chunks+1):
    chunk = chunks[chunk_num-1]
    documents.append(Document(page_content= chunk, metadata={"source": "local"}))

print(len(documents))
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
```

```python
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
docsearch = Chroma.from_documents(documents, embedding_function)
```

```python
vector_search_model = "mixtral-8x7b-32768"
def run_vector_search(relevant_lines,traceback_info):
    relevant_code = '\n\n---------------------------------\n\n'.join([doc.page_content for doc in relevant_lines])

    messages = [
        {
            "role": "system",
            "content": """You are a smart assistant. Based on the user context,
                        provide a detailed explanation of the error and suggest a solution to fix it""",
        },
        {
            "role": "user",
            "content":  f"""An error has occurred in the function with the code snippet: {relevant_code},
                            and the traceback details are: {traceback_info}. Analyze this information
                            and provide a solution or suggest necessary changes to resolve the issue.""",
        }
    ]
    response = client.chat.completions.create(
        model= vector_search_model,
        messages=messages,
        max_tokens=4096
    )
    response_message = response.choices[0].message

    return response_message.content
```

```python
def word_wrap(string, n_chars=72):
    if len(string) < n_chars:
        return string
    else:
        return string[:n_chars].rsplit(' ', 1)[0] + '\n' + word_wrap(string[len(string[:n_chars].rsplit(' ', 1)[0])+1:], n_chars)
```
